### Heap Sort in Medical Application
In emergency departments, to prioritize patients based on the urgency of their condition.

```python
#Insertion
def insert_min(heap_array, data, priority):
    if len(heap_array)==0:
        heap_array.append(data)
    else:
        heap_array.append(data)
        return reheap_up_min(heap_array, find_index_with_key(heap_array, priority), priority)
#Reheap Up    
def reheap_up_min(heap_array, index_data, data):
    if index_data == 0:
        if data < int(list(heap_array[index_data].keys())[0]):
            heap_array[index_data], heap_array[heap_array.index(find_index_with_key(heap_array, data))] = data, heap_array[index_data]
    else:
        if index_data % 2 != 0:    #i = (left - 1)//2
            ind_parent = (index_data - 1) // 2
            if data < int(list(heap_array[ind_parent].keys())[0]):
                temp = heap_array[ind_parent]
                heap_array[ind_parent] = heap_array[find_index_with_key(heap_array, data)]
                heap_array[index_data] = temp
            return reheap_up_min(heap_array, ind_parent, data)
        if index_data % 2 == 0:     #i = (right - 2)//2
            ind_parent = (index_data - 1) // 2
            if data < int(list(heap_array[ind_parent].keys())[0]):
                temp = heap_array[ind_parent]
                heap_array[ind_parent] = heap_array[find_index_with_key(heap_array, data)]
                heap_array[index_data] = temp
            return reheap_up_min(heap_array, ind_parent, data)
```

```python
def find_index_with_key(array, key):
    for i, item in enumerate(array):
        if key in item:
            return i
    return -1
```

```python
#Display elements
def delete_node_min(heap_array_sort, priority):
    n = len(heap_array_sort)
    if n > 1 : 
        heap_array_sort[0] = heap_array_sort[-1]
        heap_array_sort.pop()
        reheap_down_min(heap_array_sort)
    if n == 1:
        heap_array_sort.pop()
        return heap_array_sort       
#Reheap Down
def reheap_down_min(heap_array_sort):
    n = len(heap_array_sort)
    for i in range(len(heap_array_sort)):
        left_child = 2 * i + 1
        right_child = 2 * i + 2
        if left_child > n:
            continue
        elif left_child < n and int(list(heap_array_sort[i].keys())[0]) > int(list(heap_array_sort[left_child].keys())[0]):
            if right_child < n and int(list(heap_array_sort[i].keys())[0]) > int(list(heap_array_sort[right_child].keys())[0]) and int(list(heap_array_sort[left_child].keys())[0]) > int(list(heap_array_sort[right_child].keys())[0]):
                temp = heap_array_sort[i]
                heap_array_sort[i] = heap_array_sort[right_child]
                heap_array_sort[right_child] = temp
            else:
                temp = heap_array_sort[i]
                heap_array_sort[i]=heap_array_sort[left_child]
                heap_array_sort[left_child] = temp
        elif right_child < n and int(list(heap_array_sort[i].keys())[0]) > int(list(heap_array_sort[right_child].keys())[0]):
            if int(list(heap_array_sort[i].keys())[0]) > int(list(heap_array_sort[left_child].keys())[0]) and int(list(heap_array_sort[left_child].keys())[0]) > int(list(heap_array_sort[right_child].keys())[0]):
                temp = heap_array_sort[i]
                heap_array_sort[i] = heap_array_sort[right_child]
                heap_array_sort[right_child] = temp
            else:
                temp = heap_array_sort[i]
                heap_array_sort[i] = heap_array_sort[right_child]
                heap_array_sort[right_child] = temp
```

```python
stop_condition=True
heap_array=[]
sorted_array = []
while(stop_condition):
    print('Emergency Room Management\n\t1. Insert Patient Details\n\t2. Display patient details based on priority\n\t3. Exit')
    heap_input=int(input("\nEnter the operations to be applied : "))
    if heap_input==1: #Insert patient details
        new_patient = {}
        name = input("Enter patient name : ")
        reason = input("Enter patient status : ")
        priority = int(input('Enter the priority status of the patient : '))
        new_patient[priority] = {'PatientName' : name, "MedicalConditions " : reason}
        insert_min(heap_array, new_patient, priority)
    elif heap_input==2: #Display patient details based on priority
        heap_array_sort = heap_array.copy()
        n=len(heap_array_sort)
        sorted_array.clear()
        while(n):
            sorted_array.append(heap_array_sort[0])
            delete_node_min(heap_array_sort, priority)
            reheap_down_min(heap_array_sort)
            n -= 1
        print('Patient Details based on Priority : ')
        for i in range(len(sorted_array)):
            print('\tTreating patient with priority ',int(list(sorted_array[i].keys())[0]), ' --> ',list(sorted_array[i].values()))
    else:        
        stop_condition= False
```
