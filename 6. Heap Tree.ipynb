### Max Heap

```python
#Insertion
def insert(heap_array, data):
    if len(heap_array)==0:
        heap_array.append(data)
    else:
        heap_array.append(data)
        return reheap_up(heap_array, heap_array.index(data), data)

#Reheap Up    
def reheap_up(heap_array, index_data, data):
    if index_data == 0:
        if data > heap_array[index_data]:
            heap_array[index_data], heap_array[heap_array.index(data)] = data, heap_array[index_data]
    else:
        if index_data % 2 != 0:    #i = (left - 1)//2
            ind_parent = (index_data - 1) // 2
            if data > heap_array[ind_parent]:
                temp = heap_array[ind_parent]
                heap_array[ind_parent] = data
                heap_array[index_data] = temp
            return reheap_up(heap_array, ind_parent, data)
        if index_data % 2 == 0:     #i = (right - 2)//2
            ind_parent = (index_data - 1) // 2
            if data > heap_array[ind_parent]:
                temp = heap_array[ind_parent]
                heap_array[ind_parent] = data
                heap_array[index_data] = temp
            return reheap_up(heap_array, ind_parent, data)
```

```python
#Deletion
def delete_node(heap_array, data):
    n = len(heap_array)
    if n > 1 : 
        heap_array[0] = heap_array[-1]
        heap_array.pop()
        reheap_down(heap_array)
    if n == 1:
        heap_array.pop()
        return heap_array       

#Reheap Down
def reheap_down(heap_array):
    n = len(heap_array)
    for i in range(len(heap_array)):
        left_child = 2 * i + 1
        right_child = 2 * i + 2
        if left_child > n:
            continue
        elif left_child < n and heap_array[i] < heap_array[left_child]:
            if heap_array[i] < heap_array[right_child] and heap_array[left_child] < heap_array[right_child]:
                temp = heap_array[i]
                heap_array[i] = heap_array[right_child]
                heap_array[right_child] = temp
            else:
                temp = heap_array[i]
                heap_array[i]=heap_array[left_child]
                heap_array[left_child] = temp
        elif right_child < n and heap_array[i] < heap_array[right_child]:
            if heap_array[i] < heap_array[left_child] and heap_array[left_child] < heap_array[right_child]:
                temp = heap_array[i]
                heap_array[i] = heap_array[right_child]
                heap_array[right_child] = temp
            else:
                temp = heap_array[i]
                heap_array[i] = heap_array[right_child]
                heap_array[right_child] = temp
```
