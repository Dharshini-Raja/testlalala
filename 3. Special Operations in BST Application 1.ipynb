```python
class ContactBook:
    def __init__(self,key,value):
        self.key=key
        self.value=value
        self.left_child=None
        self.right_child=None
```

```python
#Insertion
def insert_contact(root, name, phone_no):
    if root is None:
        print('Contact is Inserted.')
        return ContactBook(name, phone_no)
    if name < root.key:
        root.left_child=insert_contact(root.left_child, name, phone_no)
    else:
        root.right_child=insert_contact(root.right_child, name, phone_no)
    return root
```

```python
#Searching contact Using Name 
def search_contact(root, name):
    if root is None or root.key==name:
        return root
    if name < root.key:
        return search_contact(root.left_child, name)
    else: #Name Greater than rightchild
        return search_contact(root.right_child, name)
```

```python
#Searching contact Using Number
def search_contactno(root, phone_no):
    if root:
        leftres=search_contactno(root.left_child, phone_no)
        if leftres:
            return leftres
        if root.value==phone_no:
            return root
        rightres=search_contactno(root.right_child, phone_no)
        if rightres:
            return rightres 
    return None
```

```python
#Display Contact
def display_contact(root):
    if root:
        display_contact(root.left_child)
        print(str(root.key) + " : " + str(root.value) + " ---> ", end=" ")
        display_contact(root.right_child)
```

```python
#Delete Contact
def delete_contact(root, name):
    if root is None:
        return root
    if name < root.key:
        root.left_child=delete_contact(root.left_child, name)
    elif name > root.key:
        root.right_child=delete_contact(root.right_child, name)
    else:  #  0 or 1 Child
        if root.left_child is None:
            temp=root.right_child
            root=None
            return temp
        if root.right_child is None:
            temp=root.left_child
            root=None
            return temp
        # 2 Child
        inorder_succ=root.right_child
        while(inorder_succ.left_child):
            inorder_succ=inorder_succ.left_child
        temp=inorder_succ
        root.key=temp.key
        root.value=temp.value
        root.right_child=delete_contact(root.right_child, temp.key)
    return root     
```

```python
#For finding Nth Minimum Value and Nth Maximum Value node
def order_node(root, val_list):
    if root:
        order_node(root.left_child, val_list)
        val_list.append(root.key)
        order_node(root.right_child, val_list)
```

```python
#Sum of Left and Right Subtree Separately
def sum_val(root, name):
    if root is None or root.key==name:
        return root
    if name < root.key:
        return sum_val(root.left_child, name)
    else: #Name Greater than rightchild
        return sum_val(root.right_child, name)
```

```python
#Distance between any two given nodes in terms of levels
def distance_node(root, level, name):
    if root is None:
        return 0
    elif root.key==name:
        return level
    elif name < root.key:
        return distance_node(root.left_child, level+1, name)
    else:
        return distance_node(root.right_child, level+1, name)
```

```python
stop_condition=True
root=None
while(stop_condition):
    print("\nContact Book Management\n\t1.Insert Contact\n\t2.Search Contact Using Name\n\t3.Search Contact Using Number\n\t4.Display Contact\n\t5.Delete Contact\n\t6.Nth Minimum Value Node and its Level\n\t7.Nth Maximum Value Node and its Level\n\t8.Sum of Left and Right Subtree Separately\n\t9.Distance between any two given nodes in terms of levels\n\t10.Exit")
    bst_input=int(input('Enter the operations to be applied : '))
    
    if bst_input==1: #Insert Contact
        name=input('\nEnter contact name to be inserted : ')
        phone_no=int(input('Enter contact number to be inserted : '))
        root=insert_contact(root, name, phone_no)
    
    elif bst_input==2: #Search Contact Using Name
        name=input('Enter contact name to be searched : ')
        searched=search_contact(root, name)
        if searched:
            print('Contact Exist and his/her contact number is ', searched.value)
        else:
            print("Contact Doesn't Exist")
    
    elif bst_input==3: #Search Contact Using Number
        phone_no=int(input('Enter contact number to be searched : '))
        searched=search_contactno(root, phone_no)
        if searched:
            print('Contact Exist and his/her contact name is ', searched.key)
        else:
            print("Contact Doesn't Exist")
    
    elif bst_input==4: #Display Contact
        display_contact(root)
    
    elif bst_input==5: #Delete Contact
        name=input('\nEnter contact name to be deleted : ')
        searched=search_contact(root, name)
        if searched:
            root=delete_contact(root, name)
            searched2=search_contact(root, name)
            if searched2 is None:
                print('Contact is deleted')
            else:
                print('Contact is not deleted!!! Try Again!!!')
        else:
            print("Contact doesn't exist")
    
    elif bst_input==6: #Nth Minimum Value Node and its Level
        min_val_list=[]
        order_node(root, min_val_list)
        min_val=int(input('Enter the nth value : '))
        if min_val>0 and min_val<=len(min_val_list):
            distance=distance_node(root, 0, min_val_list[min_val-1])
            ordinal = lambda n: "%d%s" % (n,"tsnrhtdd"[(n//10%10!=1)*(n%10<4)*n%10::4])
            print(ordinal(min_val),' Minimum value Node is ', min_val_list[min_val-1], ' and its level is ', distance)
        else:
            print('Given input of nth value exceeds the levels of the tree')
    elif bst_input==7: #Nth Maxmimum Value Node and its Level
        max_val_list=[]
        order_node(root, max_val_list)
        max_val=int(input('Enter the nth value : '))
        if max_val>0 and max_val<=len(max_val_list):
            distance=distance_node(root, 0, max_val_list[-max_val])
            ordinal = lambda n: "%d%s" % (n,"tsnrhtdd"[(n//10%10!=1)*(n%10<4)*n%10::4])
            print(ordinal(max_val),' Maximum value Node is ', max_val_list[-max_val], ' and its level is ', distance)
        else:
            print('Given input of nth value exceeds the levels of the tree')
    
    elif bst_input==8: #Sum of Left and Right Subtree Separately
        val_list=[]
        order_node(root, val_list)
        
        #Sum of Left Subtree
        index = val_list.index(root.key)
        left_list=val_list[:index]
        sum_left_list=[]
        for i in left_list:
            res=sum_val(root, i)
            sum_left_list.append(res.value)
        print('Sum of Left Subtree is ', sum(sum_left_list))
        
        #Sum of Right Subtree
        index = val_list.index(root.key)
        right_list=val_list[index+1:]
        sum_right_list=[]
        for i in right_list:
            res=sum_val(root, i)
            sum_right_list.append(res.value)
        print('Sum of Right Subtree is ', sum(sum_right_list))
    
    elif bst_input==9: #Distance between any two given nodes in terms of levels
        choi=True
        node1=input('Enter contact name 1 : ')
        searched=search_contact(root, node1)
        if searched:
            dist1=distance_node(root, 0, node1)
        else:
            print("Node Doesn't Exist!!! Couldn't execute!! Try Again")
            choi=False
        node2=input('Enter contact name 2 : ')
        searched=search_contact(root, node2)
        if searched:
            dist2=distance_node(root, 0, node2)
        else:
            print("Node Doesn't Exist!!! Couldn't execute!! Try Again")
            choi=False
        if choi:
            distance=dist1-dist2
            print('Distance between contact name 1 and contact name 2 is ', abs(distance))
    else:
        stop_condition=False
```
